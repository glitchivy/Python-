Sort array using Quick Sort

Error handling for empty array

Step-by-step pivot placement shown

```python
def partition(arr,low,high):
    pivot=arr[high]
    i=low-1
    for j in range(low,high):
        if arr[j]<pivot:
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[high]=arr[high],arr[i+1]
    return i+1 
def quick_sort(arr,low,high,show_steps=True):
    if low<high:
        pi=partition(arr,low,high)
        if show_steps:
            print(f"pivot placed at index {pi}:{arr}")
        quick_sort(arr,low,pi-1,show_steps)
        quick_sort(arr,pi+1,high,show_steps)
    return arr 
def menu():
    arr = []
    while True:
        print("\n=== Quick Sort Project Menu ===")
        print("1. Enter a new array")
        print("2. Display current array")
        print("3. Sort array (Ascending) using Quick Sort")
        print("4. Find maximum and minimum element")
        print("5. Search for an element")
        print("6. Insert an element at a position")
        print("7. Delete an element")
        print("8. Exit")
        choice=int(input("enter the operation btw 1 to 9"))
        if choice==8:
            print("thank you for using quick sort")
            break
        elif choice==1:
            arr=list(map(int,input("enter the values").split()))
            print("Values stored")
        elif choice==2:
            if arr:
                print("The current array is:",arr)
            else:
                print("LIST IS EMPTY")
        elif choice==3:
            if arr:
                print("OG ARRAY:",arr)
                quick_sort(arr,0,len(arr)-1)
                print("SORTED ARRAY:",arr)
            else:
                print("LIST IS EMPTY") 
        elif choice==4:
            print(f"Maximum of array is {max(arr)} and Minimum of array is {min(arr)}")
        elif choice==5:
            if arr:
                x=int(input("enter the element you want to find "))
                if x in arr:
                    print(f"The element {x} is found in {arr.index(x)}")
                else:
                    print("CANT FIND")
        elif choice==6:
            if arr:
                x = int(input("Enter element to insert: "))
                pos = int(input(f"Enter position (0 to {len(arr)}): "))
                if 0 <= pos <= len(arr):
                    arr.insert(pos, x)
                    print("Updated array:", arr)
                else:
                    print("Invalid position!")
            else:
                print("Array is empty!")
        elif choice==7:
             if arr:
                x = int(input("Enter element to delete: "))
                if x in arr:
                    arr.remove(x)
                    print("Updated array:", arr)
                else:
                    print("Element not found.")
            

    

        
menu()

            
            
```
