###  MULTIPLICATION TABLE GENERATOR

ðŸ”¹ Features

1.Generate Table of a Single Number â€“ Enter one number and get its table (default up to 10 or 12).

2.Generate Tables for a Range of Numbers â€“ Print multiplication tables for all numbers in a given range.

3.Custom Length Table â€“ Choose how many multiples you want (e.g., up to 20, 50, etc.).

4.Save Table(s) to File â€“ Export the generated table(s) into a text file.

5.Display All Generated Tables (History) â€“ Show all tables created in the current session.

6.Clear History â€“ Remove all previously generated tables from memory.

7.Exit â€“ Quit the program.

```python
print("MULTIPLICATION TABLE GENERATOR")
print("1.Generate Table of a Single Number")
print("2.Generate Tables for a Range of Numbers")
print("3.Custom Length Table")
print("4.Save Table(s) to File")
print("5.Display All Generated Tables")
print("6.Clear History")
print("7.Exit")
history=[]
while True:
    choice=int(input("enter the number of the operation you want to perform"))
    if choice==7:
        print("Thank you for using")
        break
    elif choice==1:
        number=int(input("enter the number you want the table for"))
        for i in range(1,11):
            multi=number*i
            table=[f"{number}x{i}={multi}"]
            history.append(table)
            print(table)
    elif choice==3:
        number=int(input("enter the number you want the table for"))
        num=int(input("enter the numbers"))
        for i in range(1,num+1):
            multi=number*i
            table=[f"{number}x{i}={multi}"]
            history.append(table)
            print(table)
    elif choice==2:
        number=int(input("enter the number you want the table for"))
        for j in range(1,number+1):
            print(f"for number {j}")
            for i in range(1,11):
                multi=j*i
                table=[f"{j}x{i}={multi}"]
                history.append(table)
                print(table)
    elif choice == 4:  # Save to file
        if not history:
            print(" No tables generated yet to save.")
        else:
            with open("tables.txt", "w") as f:
                for table in history:
                    f.write("\n".join(table))
                    f.write("\n\n")
            print(" Tables saved to 'tables.txt'")
    elif choice==6:
        history.clear()
        print("History deleted")
    elif choice == 5:  # Show history
        if not history:
            print("No tables generated yet.")
        else:
            for table in history:
                
                print("\n".join(table))

        
        
            
```
