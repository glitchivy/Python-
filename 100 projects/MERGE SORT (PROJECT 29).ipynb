### MERGE SORT PROJECT

Enter a new array – Input numbers to work with.

Display array – Shows the current array.

Sort array (ascending) – Performs Merge Sort with step-by-step splitting & merging.

Find max & min – Displays the largest and smallest elements.

Search element – Finds an element and shows its index.

Insert element – Adds an element at a specific position.

Delete element – Removes an element from the array.

Reverse array – Reverses the list in place.

```python
def merge_sort(arr,steps=True):
    if len(arr)>1:
        mid=len(arr)//2
        left=arr[:mid]
        right=arr[mid:]
        if steps:
            print(f"\n splitting steps: {arr}= {left},{right}")
        merge_sort(right)
        merge_sort(left)
        i=j=k=0
        while i<len(left) and j<len(right):
            if left[i]<right[j]:
                arr[k]=left[i]
                i+=1
            else:
                arr[k]=right[j]
                j+=1
            k+=1
        while i<len(left):
            arr[k]=left[i]
            i+=1
            k+=1
        while j<len(right):
            arr[k]=right[j]
            j+=1
            k+=1
        if steps:
            print(f"Merged{arr}")
    return arr

def menu():
    arr = []
    while True:
        print("\n=== Merge Sort Project Menu ===")
        print("1. Enter a new array")
        print("2. Display current array")
        print("3. Sort array (Ascending) using Merge Sort")
        print("4. Find maximum and minimum")
        print("5. Search for an element")
        print("6. Insert an element at a position")
        print("7. Delete an element")
        print("8. Reverse the array")
        print("9. Exit")

        choice = int(input("\nEnter your choice (1-9): "))
        if choice==9:
            print("Thank you for using it")
            break 
        elif choice==1:
            arr=list(map(int,input("enter the array you want for sorting").split()))
            print("Array stored")
        elif choice==2:
            print("stored array:")
            print(arr)
        elif choice==3:
            merge_sort(arr)
            print("The Sorted array:",arr)
        elif choice==4:
            print(f"The maximum value is {max(arr)} and the min value is {min(arr)}")
        elif choice==5:
            x=int(input("enter the element you want to find"))
            if x in arr:
                print(f"The element {x} is in {arr.index(x)}")
            else:
                print("cant find ")
        elif choice==6:
            if arr:
                x = int(input("Enter the element to insert: "))
                position = int(input(f"Enter the position (0 to {len(arr)}): "))
                if 0 <= position <= len(arr):
                    arr.insert(position, x)
                    print("Element inserted successfully!")
                    print("Updated array:", arr)
                else:
                    print("Invalid position!")
            else:
                print("Array is empty!")
        elif choice == 7:
            if arr:
                x = int(input("Enter element to delete: "))
                if x in arr:
                    arr.remove(x)
                    print("Element deleted successfully!")
                    print("Updated array:", arr)
                else:
                    print("Element not found.")
            else:
                print("Array is empty!")

        elif choice == 8:
            if arr:
                print("Earlier array:", arr)
                arr.reverse()
                print("Reversed array:", arr)
            else:
                print("Array is empty!")
if __name__ == "__main__":
    menu()
            
            
            
        
```
